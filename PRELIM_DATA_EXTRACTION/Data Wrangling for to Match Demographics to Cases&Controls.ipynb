```python
import pandas as pd    
cases = pd.read_csv('cases.csv')
cases.head()
```

```python
controls = pd.read_csv('controls.csv')
controls.head()
```

```python
admissions = pd.read_csv('ADMISSIONS.csv')
admissions.head()
```

```python
patients = pd.read_csv('PATIENTS.csv')
patients.head()
```

```python
cases_DOB = cases.merge(patients, on='SUBJECT_ID', how='left')
```

```python
cases_DOB.head()
```

```python
controls_DOB = controls.merge(patients, on='SUBJECT_ID', how='left')
```

```python
controls_DOB.head()
```

```python
cases_demo = cases_DOB.merge(admissions, on='SUBJECT_ID', how='left')
```

```python
cases_demo.head()
```

```python
controls_demo = controls_DOB.merge(admissions, on='SUBJECT_ID', how='left')
```

```python
controls_demo.head()
```

```python
controls_demo = controls_demo.drop_duplicates(subset='HADM_ID', keep="first")
```

```python
controls_demo.to_csv('controls_demo.csv', index=False)
```

```python
cases_demo = cases_demo.drop_duplicates(subset='HADM_ID', keep="first")
```

```python
cases_demo.to_csv('cases_demo.csv', index=False)
```
